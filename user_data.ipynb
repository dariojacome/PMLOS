Se busca si hay filas duplicadas

```python
# Encontrar las filas con valores duplicados 
filas_duplicadas = df[df.duplicated(subset=['user_id', 'item_id', 'playtime_forever', 'playtime_2weeks'], keep=False)]

# filas_duplicadas contendrá las filas con valores duplicados en ambas columnas 'user_id' y 'item_id'
filas_duplicadas
```

Al encontrar filas duplicadas se eliminará las filas duplicadas manteniendo la primera aparición de cada fila duplicada.

```python
# Encontrar las filas duplicadas considerando todos los valores de la fila
filas_duplicadas = df[df.duplicated(keep='first')]

# Eliminar las filas duplicadas y mantener la primera aparición
df= df.drop_duplicates(keep='first')
# se vuelve a hacer el conteo del index
df= df.reset_index(drop=True)
df
```

se extrae las columnas user_id y user_url para en un futuro ahorrar recursos en el df, asi relacionar user_url con user_id

```python
#se extrae las columnas user_id y user_url
df_user= df[['user_id','user_url']]
#se elimina duplicados de los usuarios
df_user= df_user.drop_duplicates(keep='first')
df_user= df_user.reset_index(drop=True)
df_user
```

se extrae en un csv esta lista para relacionarlos en un futuro

```python
df_user.to_csv('usuarios_id_url.csv', index=False)
```
