```python
#eliminar columnas que no se van a usar
df= df.drop(['specs','tags','early_access'], axis=1)
df
```

```python
#eliminar filas que tiene id nulo
df = df.dropna(subset=['id'])
# Reemplazar los valores nulos en la columna release_year con 0
df['release_year'] = df['release_year'].fillna(0)
df
```

Se categoriza generos para clasificar los juegos, Separando los generos en dummies para poder leerlos correctamente

```python
#se revisa los valores de los generos
# Concatenar todas las cadenas en una sola
todas_las_cadenas = ' '.join(df['genres'].dropna().astype(str))

# Dividir las cadenas en palabras individuales
palabras = todas_las_cadenas.split()

# Obtener palabras únicas
palabras_unicas = np.unique(palabras)
palabras_unicas
```

```python
df["genres"] = df["genres"].astype("category")
```

```python
## Esta funcion me ayuda a separar los generos que estan en lista y ponerlos por ,

def unir_generos(lista):
    lista = eval(lista)  # Convertir la cadena en una lista
    return ', '.join(lista)
```

```python
df["genres"] = df["genres"].apply(unir_generos) # Acá aplico la función que cree
```

```python
genres_dummies =  df["genres"].str.get_dummies(sep=", ") # Separo los generos en dummies para poder leerlos correctamente
```

```python
df = pd.concat([df, genres_dummies], axis=1) ## Los concateno a mi df
```

```python
# Borro las columnas que ya no necesito
df=df.drop(columns=['genres','nan'], axis=1)
```

```python
df.rename(columns={'id': 'item_id'},inplace=True)
```

se carga el dataframe de item con su nombre para poner nombres a los juegos de la columna title que son nulos

```python
df1=pd.read_csv('item_id_name.csv')
df1
```

se combina los dos DataFrames en función de 'item_id' y luego utiliza fillna para reemplazar los valores nulos en 'title' con los valores correspondientes de 'item_name' en caso de que haya coincidencia en 'item_id'. El resultado final estará en 'merged_df'.

```python
# Utiliza merge para combinar los DataFrames en función de 'item_id'
merged_df = df.merge(df1[['item_id', 'item_name']], on='item_id', how='left')

# Llena los valores nulos en 'title' con los valores de 'item_name'
merged_df['title'].fillna(merged_df['item_name'], inplace=True)

# Ahora, merged_df tendrá los valores de 'title' actualizados con 'item_name' si estaban nulos.
```

se evalua si funciono el codigo sumando los valores nulos de los dataframes

como se puede ver que hubo cambios en los valores nulos se procede a reemplazar la columna 'title' de merged_df a df

```python
# Reemplaza la columna 'title' en df con los valores actualizados de 'merged_df'
df['title'] = merged_df['title']
```
