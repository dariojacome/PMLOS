<h1>solucion 1 problema

```python
# importamos las columnas developer e item_id
df_dev=df_items[['developer', 'price', 'release_year']]

#borramos los valores nulos en developer
df_dev=df_dev.dropna(subset=['developer'])

df_dev=df_dev.sort_values(by='developer', ascending=True)
# Filtrar el DataFrame para excluir las filas con valores diferentes de 0 en 'price'
df_dev = df_dev[df_dev['price'] == 0]
df_dev
```

indicamos los valores de la columna developer y cuantas veces se repite

```python
df_dev['developer'].value_counts()
```

como indica muchos desarrolladores, se hace un filtro eliminando los desarrolladores que tienen 1 juego gratis

```python
valores = df_dev['developer'].value_counts() == 1

# Filtrar el DataFrame para excluir las filas con valores que se repiten solo una vez
df_dev = df_dev[~df_dev['developer'].isin(valores[valores].index)]
df_dev
```

se cuenta cuantas veces se repite cada valor de la columna developer

```python
df_dev['developer'].value_counts()
```

```python
df_dev.rename(columns={'release_year': 'Año'}, inplace=True)
```

se guarda el archivo para subir a fastapi y no use muchos recursos 

```python
df_dev.to_csv('1punto.csv', index=False)
```

se genera la solución sin fastapi

```python
def developer(desarrollador: str):
    
    # Filtrar el DataFrame para obtener solo las filas correspondientes al desarrollador especificado
    df_filtrado = df_dev[df_dev['developer'] == desarrollador]

    # Calcular la cantidad de items por año
    cantidad_items_por_anio = df_filtrado.groupby('Año').size()

    # Calcular el porcentaje de contenido gratuito por año
    porcentaje_contenido_gratuito_por_anio = (df_filtrado.groupby('Año').size())/((cantidad_items_por_anio).sum())*100

    # Crear un nuevo DataFrame con los resultados
    resultado = pd.DataFrame({'Cantidad': cantidad_items_por_anio,
                              'Contenido Free': porcentaje_contenido_gratuito_por_anio})
    
    # Formatear el porcentaje con '%' al final
    resultado['Contenido Free'] = resultado['Contenido Free'].astype(str) + '%'

    return resultado
```

```python
developer('3D Realms (Apogee Software)')
```

def userdata( User_id : str ): Debe devolver cantidad de dinero gastado por el usuario, el porcentaje de recomendación en base a reviews.recommend y cantidad de items.

Se hace un merge entre los diferentes dataframe para dar solución al 2 problema

```python
# Realiza la relación entre df_user y df_item en base a 'item_id' y agrega la columna 'price' de df_item
df_datauser = pd.merge(df_user[['user_id', 'item_id']], df_items[['item_id', 'price']], on='item_id', how='left')
df_datauser = pd.merge(df_datauser, df_review[['user_id', 'item_id', 'recommend']], on=['user_id', 'item_id'], how='left')

# la columna 'user_id' se pasara a minusculas
df_datauser['user_id'] = df_datauser['user_id'].str.lower()
df_datauser
```

```python
#se revisa los valores nulos en la columna
df_datauser['recommend'].isnull().sum()
```

```python
# Reemplaza los valores nulos en la columna 'recommend' por False
df_datauser['recommend'].fillna(False, inplace=True)
```

se borra la columna item_id para liberar espacio

```python
df_datauser = df_datauser.drop(columns=['item_id'])
```

se genera un parquet para utilizar con fastapi

```python

df_datauser.to_parquet('2punto.parquet', index=False)
```

```python
def userdata(user_id: str):
    user_id = user_id.lower()
    comprobar = list(set(df_datauser['user_id']))
    if user_id in comprobar:
           
         # Filtrar el DataFrame por el usuario especificado
        user_data = df_datauser[df_datauser['user_id'] == user_id]
        
        # Contar las repeticiones del usuario
        contador = user_data.shape[0]
        
        # Sumar la columna 'price' para las filas correspondientes al usuario
        suma_precios = user_data['price'].sum()
        # Filtrar las filas donde la columna 'recommend' es True y luego sumar esos valores para dividirlos entre el contador
        promedio = (user_data[user_data['recommend'] == True]['recommend'].sum())*100/contador


        print(f"El usuario {user_id} tiene {contador} items.")
        print(f"La suma dinero gastado para este usuario es: {suma_precios}")
        print(f"El porcentaje de recomendaciones para este usuario es: {promedio:.2f} %")
    
        return 
    else:
        print('Ingrese nuevamente el usuario')
        return 
```

<h1> Solución problema 2

def countreviews( YYYY-MM-DD y YYYY-MM-DD : str ): Cantidad de usuarios que realizaron reviews entre las fechas dadas y, el porcentaje de recomendación de los mismos en base a reviews.recommend.

```python
# Convertir la columna 'year' en una variable de tiempo
df_review['year'] = pd.to_datetime(df_review['year'], format='%Y')
```

```python

def countreviews(start_date, end_date: str):
    # Convertir las fechas de inicio y fin en objetos datetime
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)

    # Filtrar el DataFrame para obtener las revisiones en el rango de fechas
    filtered_df = df_review[(df_review['year'] >= start_date)& (df_review['year'] <= end_date)]

        
    # Contar los comentarios hechos por los usuarios
    cont = filtered_df.shape[0]  
    
    porcentaje = (filtered_df[filtered_df['recommend'] == True]['recommend'].sum())*100/cont

    print(f'Cantidad de usuarios que realizaron reviews entre las fechas {start_date} y {end_date} es de: {cont}')
    print(f"El porcentaje de juegos que fueron recomendados es: {porcentaje:.2f} %")
    
    return 
```

Se realiza una prueba ingresando dos fechas con el formato YYYY-MM-DD  YYYY-MM-DD

```python
countreviews('2011-01-01', '2012-12-31')
```

<h1> Solución problema 3

def genre( género : str ): Devuelve el puesto en el que se encuentra un género sobre el ranking de los mismos analizado bajo la columna PlayTimeForever.

```python
# Realiza la relación entre df_user y df_item en base a 'item_id' 
df_genre = pd.merge(df_user, df_items, on='item_id', how='right')
df_genre
```

Se elimina las filas que contienen nulos en la columna user_id y se convierte la columna playtime_forever a int64 para reducir espacio

```python
df_genre=df_genre.dropna(subset=['user_id'])

df_genre['playtime_forever'] = df_genre['playtime_forever'].astype('int64')

df_genre
```

Se eliminan las columnas 'publisher', 'title', 'price', 'developer', 'release_year' que no se van a usar en las preguntas 3 y 4

```python
df_genre = df_genre.drop(columns=['publisher', 'title', 'price', 'developer', 'release_year'])
df_genre
```

se borra las filas que tienen valores 0 en las columnas playtime_forever, ya que no se las usara para los puntos 3 y 4 de la funcion

```python
df_genre = df_genre.loc[~((df_genre['playtime_forever'] == 0) )]
df_genre
```

```python
# Reemplaza True por su equivalente en la columna "playtime_forever" para las de los generos para su posterior analisis
df_genre[df_genre.columns[3:]] = df_genre[df_genre.columns[3:]].multiply(df_genre['playtime_forever'], axis='index')

df_genre
```

Se realiza un diccionario con la suma de cada genero basado en playtime_forever de cda usuario y se los ordena de mayor a menor

```python
# Calcula la suma total de cada columna y crea el diccionario puestos_genre
puestos_genre = df_genre[df_genre.columns[3:]].sum().to_dict()

# Ordena el diccionario de mayor a menor según los valores
puestos_genre = dict(sorted(puestos_genre.items(), key=lambda x: x[1], reverse=True))

puestos_genre
```

Se coloca los puestos de cada genero

```python
# Valor inicial
valor_incremental = 1

# Recorrer las claves del diccionario y asignar valores incrementales
for clave in puestos_genre.keys():
    puestos_genre[clave] = valor_incremental
    valor_incremental += 1

# Imprimir el diccionario con los valores reemplazados
puestos_genre
```

Se hace la función basada para usar en la API

```python
def genre(genero: str):
    if genero in puestos_genre:
        valor = puestos_genre[genero]
        print(f"El género '{genero}' se encuentra en el puesto {valor}")
    else:
        print(f"El género '{genero}' no se encuentra en el diccionario")
```

se realiza una prueba ingresando un genero

```python
# Se ingresa el genero para saber su ranking
genre('Action')
```

def userforgenre( género : str ): Top 5 de usuarios con más horas de juego en el género dado, con su URL (del user) y user_id.

Se carga el csv que contiene el usuario su user_id con su respectivo user_url

```python
df_urluser=pd.read_csv('usuarios_id_url.csv')
df_urluser
```

```python
#se agrupa por usuario y se suma las demas columnas para obtener las horas jugadas de cada usuario por genero
df_genre_user = df_genre.groupby('user_id').sum(numeric_only=True)

#se elimina las columnas item_id y playtime_forever
df_genre_user = df_genre_user.drop(['item_id', 'playtime_forever'], axis=1)

df_genre_user
```

se fusiona (merge) entre dos DataFrames, df_genre_user y df_urluser, utilizando la columna 'user_id' como clave de fusión

```python
df_genre_user=df_genre_user.merge(df_urluser, on='user_id', how='left')
df_genre_user
```

la función encuentra y muestra el top 5 de usuarios con más horas jugadas en un género de juego específico.

```python
def userforgenre( genero : str ):
    
    #encuentra las filas de los primeros 5 valores basada en genero que ingresa
    top5 = df_genre_user.nlargest(5, genero)
    
    print(f"el top 5 de usuarios de horas jugadas segun el genero '{genero}' es:")
    print(top5[['user_id','user_url',genero]].to_string(index=False))   
```
