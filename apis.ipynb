```python
# Realiza la relación entre df_user y df_item en base a 'item_id' y agrega la columna 'price' de df_item
df_datauser = pd.merge(df_user[['user_id', 'item_id']], df_items[['item_id', 'price']], on='item_id', how='left')
df_datauser = pd.merge(df_datauser, df_review[['user_id', 'item_id', 'recommend']], on=['user_id', 'item_id'], how='left')

# la columna 'user_id' se pasara a minusculas
df_datauser['user_id'] = df_datauser['user_id'].str.lower()
df_datauser
```

```python
def userdata(user_id: str):
    user_id = user_id.lower()
    comprobar = list(set(df_datauser['user_id']))
    if user_id in comprobar:
           
        """
        Esta función cuenta cuántas veces se repite un usuario en la columna 'user_id' del DataFrame df.
        Luego suma la columna 'price' de las filas correspondientes al usuario.

        Args:
        user_id (str): El usuario que deseas contar en formato de cadena.

        Returns:
        El número de repeticiones del usuario en la columna 'user_id', la suma de la columna 'price' y
        el promedio de recomendaciones para este usuario
        """
        # Filtrar el DataFrame por el usuario especificado
        user_data = df_datauser[df_datauser['user_id'] == user_id]
        
        # Contar las repeticiones del usuario
        contador = user_data.shape[0]
        
        # Sumar la columna 'price' para las filas correspondientes al usuario
        suma_precios = user_data['price'].sum()
        # Filtrar las filas donde la columna 'recommend' es True y luego sumar esos valores para dividirlos entre el contador
        promedio = (user_data[user_data['recommend'] == True]['recommend'].sum())*100/contador


        print(f"El usuario {user_id} tiene {contador} items.")
        print(f"La suma dinero gastado para este usuario es: {suma_precios}")
        print(f"El porcentaje de recomendaciones para este usuario es: {promedio:.2f} %")
    
        return 
    else:
        print('Ingrese nuevamente el usuario')
        return 

```

```python
# Convertir la columna 'year' en una variable de tiempo
df_review['year'] = pd.to_datetime(df_review['year'], format='%Y')
```

```python

def countreviews(start_date, end_date: str):
    # Convertir las fechas de inicio y fin en objetos datetime
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)

    # Filtrar el DataFrame para obtener las revisiones en el rango de fechas
    filtered_df = df_review[(df_review['year'] >= start_date)& (df_review['year'] <= end_date)]

        
    # Contar los comentarios hechos por los usuarios
    cont = filtered_df.shape[0]  
    
    porcentaje = (filtered_df[filtered_df['recommend'] == True]['recommend'].sum())*100/cont

    print(f'Cantidad de usuarios que realizaron reviews entre las fechas {start_date} y {end_date} es de: {cont}')
    print(f"El porcentaje de juegos que fueron recomendados es: {porcentaje:.2f} %")
    
    return 
```

```python
countreviews('2011-01-01', '2012-12-31')
```
