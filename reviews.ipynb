```python
import pandas as pd
import ast
import numpy as np
from langdetect import detect
from googletrans import Translator
from textblob import TextBlob
```

```python
df= pd.read_csv('NuevoReviews.csv')
df
```

```python
#se elimina las columnas que no se van a usar
df = df.drop(['funny','last_edited','helpful'], axis=1)
df
```

```python
df['review'].isnull().sum()
```

```python
# los valores nulos en 'review' se reemplazan por cadenas de texto vacías
df['review'] = df['review'].fillna('')
```

```python
df['review'].isnull().sum()
```

Para hacer un buen analisis de sentimientos en la columna de review primero se detecta si hay mas idiomas aparte del ingles

```python
lenguajes_detectados = set()

for texto in df['review']:
    try:
        lenguaje = detect(texto)
        lenguajes_detectados.add(lenguaje)
    except:
        # En caso de que no se pueda detectar el lenguaje, continuar
        pass

# Imprimir los lenguajes detectados únicos
print("Lenguajes detectados en la columna 'review':")
for lenguaje in lenguajes_detectados:
    print(lenguaje)

```

Se detecto multiples idiomas, por lo que se hará la traducción de la columna reviews a ingles

```python
#función translate_to_english intenta traducir el texto a inglés utilizando googletrans. Si la traducción es exitosa, 
# se agrega la traducción a una nueva columna llamada 'texto_traducido'. Si la traducción falla, simplemente se mantiene 
# el texto original.
def translate_to_english(text):
    try:
        translator = Translator()
        translated_text = translator.translate(text, src='auto', dest='en')
        return translated_text.text
    except Exception as e:
        # Si la traducción falla, simplemente retorna el texto original
        return text

# Aplica la función de traducción a la columna 'review'
df['texto_traducido'] = df['review'].apply(translate_to_english)
```

Una vez traducido se procede a hacer el analisis de sentimiento segun la escala sugerida: valor '0' si es malo, '1' si es neutral y '2' si es positivo

```python
# Función para analizar el sentimiento y asignar valores según tu escala
def analyze_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity < 0:
        return 0  # Sentimiento negativo
    elif analysis.sentiment.polarity == 0:
        return 1  # Sentimiento neutral
    else:
        return 2  # Sentimiento positivo

# Aplicamos el análisis de sentimiento y creamos la columna 'sentiment_analysis'
df['sentiment_analysis'] = df['texto_traducido'].apply(analyze_sentiment)

# Si la reseña está ausente, asignamos el valor 1 (neutral) como predeterminado
df['sentiment_analysis'].fillna(1, inplace=True)
```

una vez hecho el analisis de sentimiento se borra las clumnas review y texto_traducido

```python
# Borro las columnas que ya no necesito
df=df.drop(['review','texto_traducido'], axis=1)
df
```

```python
# Separamos la columna "posted" en varias columnas utilizando el método split()

posted = df["posted"].str.split(expand=True)

# Damos nombres a las nuevas columnas resultantes.
posted.columns = ["posted", "month", "day", "year"]

# Concatenamos las nuevas columnas al DataFrame original.
df = pd.concat([df, posted], axis=1)
df
```

```python
# Borramos el . que quedó del "year"
df["year"] = df["year"].str.removesuffix(".")
df
```

```python
df["year"].fillna(2023, inplace=True)
```
